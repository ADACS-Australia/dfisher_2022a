### A grand function to include all the procedures


```python
model
```

```python
def grand(cube, model, line, z=0, fit_method='leastsq', snr_threshold=5., left=15, right=15):
    p = ProcessedCube(cube, z)
    ss = SNRMap(cube, threshold=snr_threshold)
    fline = Line(line)
    cr = CubeRegion(cube, fline, snr_threshold=snr_threshold, right=right, left=left)
    cfl = CubeFitterLM(data=p.data, weight=p.weight, model=model, x=p.x, fit_method=fit_method)
    return cfl.result
    
```

```python
def grand_prepare_cube(cube, line, z=0, snr_threshold=5., left=15, right=15):
    out = ResultLM()
    p = ProcessedCube(cube,z)
    out.z = p.z
    fline = Line(line)
    out.line = line
    out.snr_threshold = snr_threshold
    cr = CubeRegion(p.cube, fline, snr_threshold=snr_threshold, right=right, left=left)
    cu = cr.get_regional_cube()
    p.cube = cu
    return out
```
